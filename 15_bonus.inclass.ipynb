## Regresión lineal con scikit-learn

La principal librería de Machine Learning en general para Python: 

https://scikit-learn.org/stable/

Nos hace muy fácil hacer desde regresiones lineales a ensembl models como random forests.

Nos hacemos datos de juguete que siguen aproximadamente la recta `y = 1.32x + 30`:

````python
xs = np.random.normal(40, 20, 1000)
ys = xs * 1.32 + 30 + np.random.normal(scale=25, size=1000)

plt.scatter(xs, ys, alpha=.4)
````

Hacer una regresión lineal es tan fácil como:

````python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(xs.reshape(-1,1), ys) # reshape es necesario porque fit espera un vector columna
````

Para dibujar la recta que hemos encontrado podemos hacer uso de su pendiente y su coordenada en el origen:

````python
(lr.coef_, lr.intercept_)
````

````python
plt.scatter(xs, ys, alpha=.4)
plt.plot(xs, xs * lr.coef_ + lr.intercept_, c='Grey')
````

Pero es más genérico usar `predict`, que está implementado para __todos__ los modelos en scikit-learn:

````python
lr.predict(20)
````

````python
np.linspace(-20, 100, 13)
````

````python
lr.predict(np.linspace(-20, 100, 13).reshape(-1, 1))
````

````python
plt.scatter(xs, ys, alpha=.4)
predictions = lr.predict(xs.reshape(-1, 1))
plt.plot(xs, predictions, c='Grey')
````

De este modo, podemos sustituir nuestra regresión lineal por cualquier otro modelo, si nuestros datos tienen un aspecto más complicado:

````python
curvy_ys = ys = xs ** 3 + 30 + np.random.normal(scale=25e3, size=1000)
plt.scatter(xs, curvy_ys, alpha=.2)
````

... y ajustar un modelo más elaborado será exactamente igual de (poco) trabajoso que ajustar la anterior regresión lineal! Tan fácil como `.fit()` seguido de `.predict()`!

````python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(xs.reshape(-1, 1), curvy_ys)
````

````python
sorted(predictions)
````

````python
predictions = rf.predict(xs.reshape(-1, 1))
plt.scatter(xs, curvy_ys, alpha=.2)
plt.plot(sorted(xs), sorted(predictions), c='Grey')
````

#### Ejercicio

Escribe una funcion `how_far_from_iss` que tome longitud y latitud y devuelva la distancia en grados desde ese punto a la iss en el momento de llamar a la función.

````python
import math
import json
import requests

def how_far_from_iss(longitude, latitude):
    response = requests.get('http://api.open-notify.org/iss-now.json')
    dictionary = json.loads(response.content)
    iss_lon = float(dictionary['iss_position']['longitude'])
    iss_lat = float(dictionary['iss_position']['latitude'])
    
    distance = math.sqrt(iss_lon ** 2 + iss_lat **2)
    return distance
    

how_far_from_iss(-3, 40)
````
